```python
import pandas as pd
import json
import requests
from pprint import pprint
from config import polygon_API_Key as key
import time
```

```python
df = pd.read_csv('./sp500list.csv')
```

```python
 tckr = df['Symbol'].tolist()
```

```python
price_info= {'Ticker':[], 'Closing Price1':[], 'Closing Price2':[]}
```

```python
# This takes +- 2.5 hours to run
for i in tckr:
    try:
        request_url= f'https://api.polygon.io/v2/aggs/ticker/{i}/range/2/day/2021-04-19/2021-04-21?apiKey={key}'
        ticker_raw = requests.get(request_url)
        tckr_info = ticker_raw.json()
#         print(request_url)
        
#         pprint(tckr_info)
        price_info['Ticker'].append(i)
        price_info['Closing Price1'].append(tckr_info['results'][0]['c'])
        price_info['Closing Price2'].append(tckr_info['results'][1]['c'])
        time.sleep(13)

    except:
        print(f'result error'{price_info['Ticker']})
        pass
        
```
